## **How to Get Started with CrewAI Agent Collaboration**

*CrewAI supports several agentic workflow management through processes and collaboration. Let's take a look at how to implement each in a crew.*

---

**Overview on Collaboration**

Collaboration in CrewAI enables agents to work together as a team by delegating tasks and asking questions to leverage each other’s expertise. When ***allow_delegation=True***, agents automatically gain access to powerful collaboration tools.

*More about this in our [docs](https://docs.crewai.com/en/concepts/collaboration#collaboration).*

**Overview on Processes**

Processes orchestrate the execution of tasks by agents, akin to project management in human teams. These processes ensure tasks are distributed and executed efficiently, in alignment with a predefined strategy.

*   **Sequential**: Executes tasks sequentially, ensuring tasks are completed in an orderly progression.
*   **Hierarchical**: Organizes tasks in a managerial hierarchy, where tasks are delegated and executed based on a structured chain of command. A manager language model (**`manager_llm`**) or a custom manager agent (**`manager_agent`**) must be specified in the crew to enable the hierarchical process, facilitating the creation and management of tasks by the manager.

*More about this in our [docs](https://docs.crewai.com/en/concepts/processes).*

### **Collaborative Single Task**

---

**Agents**
1.   Researcher: searches the web for recent content
2.   Writer: uses the content from the researcher agent to write an article
3.   Editor: edits the article written by the writer

**Task**

*   Article Task: all the 3 agents collaborate to write one article. They delegate work between themselves using the ***allow_delegation=True*** parameter.

**Delegate Work Tool**

* Delegate work to coworker(**task: str, context: str, coworker: str**)

**Ask Question Tool**

* Ask question to coworker(**question: str, context: str, coworker: str**)

```python
from crewai import Agent, Crew, Task, Process, LLM
from crewai_tools import EXASearchTool
from IPython.display import display, Markdown

# Initialize LLM and tools
llm = LLM(model="gpt-5", api_key=OPENAI_API_KEY)
tool = EXASearchTool(api_key=EXA_API_KEY)

# Create collaborative agents focused on multi-technology AI analysis
researcher = Agent(
    role="Advanced AI Technologies Research Specialist",
    goal="Research and analyze cutting-edge AI technologies including Agentic AI, MCP, Agent Communication Protocols, RAG, Multi-Agent Systems, Vector Databases, and A2A frameworks",
    backstory="""You're a leading AI research specialist with deep expertise in analyzing
    multiple interconnected AI technologies. You excel at understanding how Agentic AI,
    Model Context Protocol (MCP), IBM's Agent Communication Protocol, Retrieval Augmented
    Generation (RAG), Multi-Agent Systems, Vector Databases, and Google's A2A framework
    work together to create powerful AI solutions. You specialize in finding technical
    documentation, implementation case studies, and performance benchmarks across these technologies.""",
    allow_delegation=True,
    llm=llm,
    tools=[tool],
    verbose=True
)

writer = Agent(
    role="AI Technology Integration Journalist",
    goal="Create comprehensive analysis of how multiple AI technologies work together to transform domain projects",
    backstory="""You're an expert technology journalist who specializes in explaining complex
    AI technology stacks and their synergistic effects. You excel at demonstrating how
    technologies like Agentic AI, MCP, Agent Communication Protocols, RAG, Multi-Agent Systems,
    Vector Databases, and A2A frameworks complement each other to solve real-world challenges.
    Your writing connects technical capabilities to business outcomes with the clarity and
    authority of premier technology journalism.""",
    allow_delegation=True,
    llm=llm,
    verbose=True
)

editor = Agent(
    role="AI Technology Stack Editor",
    goal="Ensure comprehensive coverage and technical accuracy across multiple AI technology domains",
    backstory="""You're a senior technology editor with expertise in AI technology ecosystems.
    You ensure articles comprehensively cover the interconnections between Agentic AI, MCP,
    Agent Communication Protocols, RAG, Multi-Agent Systems, Vector Databases, and A2A frameworks.
    You focus on technical accuracy while maintaining clarity about how these technologies
    collectively help organizations achieve superior outcomes in their domain projects.""",
    allow_delegation=True,
    llm=llm,
    verbose=True
)

# Create a task that encourages collaboration
article_task = Task(
    description="""Write a comprehensive 2000-word investigative article about {topic} for {year}.
    
    The article should explore:
    - Current state of Agentic AI adoption across major industries (healthcare, finance, manufacturing, retail, etc.)
    - Specific use cases and success stories with measurable impacts
    - How autonomous agents are solving complex domain-specific challenges
    - Technical capabilities that make Agentic AI transformative (reasoning, planning, tool use, collaboration)
    - Economic and productivity benefits organizations are experiencing
    - Challenges and limitations in current implementations
    - Future potential and emerging trends
    - Expert opinions and industry perspectives
    
    Structure it like a New York Times technology feature article with:
    - Compelling headline and lede
    - Real-world examples and case studies
    - Expert quotes and industry analysis
    - Balanced perspective on benefits and challenges
    - Forward-looking conclusion
    
    The byline author should be 'CrewAI Agent and Anup Kumar'.
    Collaborate with your teammates to ensure both technical accuracy and compelling storytelling.""",
    expected_output="A well-researched, engaging 2000-word investigative article with proper NYT structure, real-world examples, expert insights, and comprehensive citations.",
    agent=writer  # Writer leads, but can delegate research to researcher
)

# Create collaborative crew
crew = Crew(
    agents=[researcher, writer, editor],
    tasks=[article_task],
    process=Process.sequential,
    verbose=True
)

result = crew.kickoff(inputs={
    "topic": "The Impact of Agentic AI Across Different Domain Projects: How Autonomous Agents Are Revolutionizing Industries", 
    "year": "2025"
})
```

### **Sequential Collaboration**

---

Task execution follows the predefined order in the task list, with the output of one task serving as context for the next.
To customize task context, utilize the context parameter in the Task class to specify outputs that should be used as context for subsequent tasks.


```python
researcher = Agent(
    role="Research Specialist",
    goal="Find accurate, up-to-date information on any topic",
    backstory="""You're a meticulous researcher with expertise in finding
    reliable sources and fact-checking information across various domains.
    Use the tool to do in-depth research.
    """,
    allow_delegation=False,
    llm=llm,
    tools=[tool],
    verbose=True
)

writer = Agent(
    role="Content Writer",
    goal="Create engaging, well-structured content",
    backstory="""You're an experienced New York Times article writer who excels at transforming
    research content into compelling, readable content for different audiences. in the style of NYT articles.""",
    allow_delegation=False,
    llm=llm,
    verbose=True
)

editor = Agent(
    role="Content Editor",
    goal="Ensure content quality and consistency",
    backstory="""You're an experienced editor with an eye for detail,
    ensuring content meets high standards for clarity and accuracy of NYT articles.""",
    allow_delegation=False,
    llm=llm,
    verbose=True
)

researcher_task = Task(
    description="Research the topic {topic} for this {year} in detail.",
    expected_output="A detailed research report with citations.",
    agent=researcher
)

writer_task = Task(
    description="""Write a comprehensive 2000-word article about {topic} for this {year}.
    The article should be structured like a New York Time article. Make sure the byline author is
    this 'CrewAI Agent and Anup Kumar'.""",
    expected_output="A well-researched, engaging 2000-word article with proper structure and citations.",
    agent=writer,
    context=[researcher_task]
)

editor_task = Task(
    description="Edit the article to ensure proper structure like a New York Times article.",
    expected_output="A well-researched, engaging 2000-word article with proper structure and citations like a NTY article.",
    agent=editor,
    context=[researcher_task, writer_task]
)


# Create collaborative crew
crew = Crew(
    agents=[researcher, writer, editor],
    tasks=[researcher_task, writer_task, editor_task],
    process=Process.sequential,
    verbose=True
)

# Run the crew
result = crew.kickoff(inputs={"topic": "Biggest Private Equity deals", "year":"2025"})
```

```python
Markdown(result.raw)
```

### **Hierarchical Collaboration**

---

For complex projects, use a hierarchical process with a manager agent. Always start with sequantial then introduce hierarchical when you need to.

```python
# Manager agent coordinates the team
manager = Agent(
    role="Project Manager",
    goal="Coordinate team efforts and ensure comprehensive analysis of MCP implementations",
    backstory="Experienced project manager skilled at delegation and quality control with expertise in AI/ML technologies",
    allow_delegation=True,
    llm=llm,
    verbose=True
)

# Specialist agents
researcher = Agent(
    role="Technical Researcher",
    goal="Provide accurate research and analysis on Model Context Protocol implementations",
    backstory="Expert researcher specializing in AI protocols, API integrations, and technical architecture analysis",
    tools=[tool],
    allow_delegation=False,  # Specialists focus on their expertise
    verbose=True
)

writer = Agent(
    role="Technical Writer",
    goal="Create compelling technical content and use case documentation",
    backstory="Skilled technical writer who excels at creating comprehensive protocol analysis reports for engineering leaders and decision makers.",
    allow_delegation=False,
    verbose=True
)

# Manager-led task
project_task = Task(
    description="""Create a comprehensive analysis report examining {topic} across {use_case_scope}.
    You have authority to delegate tasks to the specialists. Focus on:
    - Current implementation patterns and best practices
    - Real-world use cases and adoption scenarios
    - Technical advantages and limitations
    - Integration challenges and solutions
    - Future trends and recommendations
    
    To ensure accuracy, make sure you review the final content before it's ready for publication. 
    Delegate research and writing tasks until the report is comprehensive and publication-ready.
    """,
    expected_output="Executive summary, detailed technical analysis, use case documentation, implementation examples, and strategic recommendations with cited sources.",
    agent=manager  # Manager will delegate to specialists
)

# Hierarchical crew
crew = Crew(
    agents=[manager, researcher, writer],
    tasks=[project_task],
    process=Process.hierarchical,  # Manager coordinates everything
    manager_llm=llm,  # Specify LLM for manager
    verbose=True
)

result = crew.kickoff(inputs={
    "topic": "Model Context Protocol usage", 
    "use_case_scope": "different use cases including development tools, data analysis, automation, and enterprise integrations"
})
```

### **Best practices for agent collaboration**

---

*   **Clear role definitions**: be specific and make sure you have complementary roles for your crew.
*   **Strategic delegation**: enable delegation for coordinators and generalists. Disable for focused specialists (optional).
*   **Context sharing**: use context sharing parameter for task dependencies.
*   **Clear task descriptions**: specific, actionable descriptions. Vague or general task description fail to guide collaboration leading to mediocre outputs.

**Next Steps:**

*   **Try the examples**: Start with the basic collaboration example
*   **Experiment with roles**: Test different agent role combinations
*   **Monitor interactions**: Use **`verbose=True`** to see collaboration in action
*   **Optimize task descriptions**: Clear tasks lead to better collaboration
*   **Scale up**: Try hierarchical processes for complex projects
